```python
import numpy as np
import pandas as pd
import os
```

```python
def Collect_dataset_per_participant(participant_no, sampling_value):
    
    data_test = pd.read_csv('Extracted_data/P' + str(participant_no) + '.csv')
    
    df_timestamp_only = pd.DataFrame(data=np.arange(start=0, stop=150, step=0.01), columns=["timestamp"]).round(2)
    data_test.rename(columns={'baseline-timestamp': 'timestamp', 
                              'Nback_LL-timestamp': 'timestamp',
                              'Nback_ML-timestamp': 'timestamp', 
                              'Nback_HL-timestamp': 'timestamp',
                              'VIS_LL-timestamp': 'timestamp', 
                              'VIS_ML-timestamp': 'timestamp',
                              'VIS_HL-timestamp': 'timestamp'}, inplace= True)
    d = []
    for i in range(7):
        df1 = df_timestamp_only
        df2 = data_test.iloc[:, i*6:(i+1)*6]
        u = df1.assign(Cnt=df1.groupby('timestamp').cumcount())
        v = df2.assign(Cnt=df2.groupby('timestamp').cumcount())
        temp = (pd.merge(u,v, on=['timestamp', 'Cnt'], how='left')).drop('Cnt', 1)
        assert len(temp) == 15000
        d.append(temp)

    df_final = pd.concat(d,axis=1)
    df_final = df_final.loc[:,~df_final.columns.duplicated()]
    df_final.ffill(inplace=True)
    df_final.bfill(inplace=True)

#     #removing the first and last x "seconds_to_remove" seconds, do not remove when switching to real tim
#     seconds_to_remove = 20

#     # df_final = df_final.copy()

#     df_final = df_final.loc[df_final['timestamp'] < (df_final['timestamp'].iloc[-1] - seconds_to_remove)]
#     df_final = df_final.loc[df_final['timestamp'] > (df_final['timestamp'].iloc[0] + seconds_to_remove)]
#     df_final.reset_index(drop=True, inplace = True)
#     df_final['timestamp'] = df_final['timestamp'] - df_final.iloc[0]['timestamp']


    #selecting the specified time (time_duration), unfortunately always start from the beginning
    time_duration = 90

    df_final = df_final.loc[df_final['timestamp'] < (df_final['timestamp'].iloc[0] + time_duration)]
    df_final.reset_index(drop=True, inplace = True)
    df_final = df_final[::sampling_value]
    df_final.reset_index(drop= True,inplace =True)
    df_final.insert(0, 'Participant', participant_no)
    appended_datapoints.append(df_final)

    
#     return df_final
```

```python
fixed_path_file_eye_data = '../../../Apx-HMI_study_data/recordings/'
participants_folders = os.listdir(fixed_path_file_eye_data)
pariticpants_list = []
sampling = 10 #take the data every 100 msec instead of 10 msec (default)
for partic in range(len(participants_folders)):
    pariticpants_list.append(int(participants_folders[partic][1:]))
pariticpants_list = sorted(pariticpants_list) 
appended_datapoints = []
print(pariticpants_list)

# one participant for debuging, uncomment the return line of the function
# df = Collect_dataset_per_participant(pariticpants_list[0],sampling)
# df


for i in pariticpants_list:
    print(i)
    __ = Collect_dataset_per_participant(i,sampling)

appended_datapoints = pd.concat(appended_datapoints)
appended_datapoints.reset_index(drop=True, inplace = True)
appended_datapoints.to_csv('appended_datapoints_all_data.csv',index =False)
```
